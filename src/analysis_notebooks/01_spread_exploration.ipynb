# Imports

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parents[1]    
sys.path.append(str(ROOT))
from src.data.refinitiv_client import get_price_timeseries
import matplotlib.pyplot as plt
```

```python
pltr_df = get_price_timeseries(
    ["PLTR.O"],
    start_date="2025-01-01",
    end_date="2025-12-23",
    interval="daily",
)

pltr_ohlc = pltr_df.copy()
pltr_ohlc.columns = ["Open", "High", "Low", "Close"]  # since it's just one ticker
pltr_ohlc = pltr_ohlc.reset_index().rename(columns={"Date": "Date"})
```

#

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
        go.Candlestick(
            x=pltr_ohlc["Date"],
            open=pltr_ohlc["Open"],
            high=pltr_ohlc["High"],
            low=pltr_ohlc["Low"],
            close=pltr_ohlc["Close"],
        )
    ]
)

fig.update_layout(
    title="PLTR.O Daily Candlestick",
    xaxis_title="Date",
    yaxis_title="Price",
    xaxis_rangeslider_visible=False,
)
fig.show()
```
